**W.1 Pandas Practice** 

Create a pandas series from each of the items below: a list, numpy and a dictionary

mylist = list('abcedfghijklmnopqrstuvwxyz')

myarr = np.arange(26)

mydict = dict(zip(mylist, myarr))

```python
import pandas as pd
import numpy as np


char_list = list('abcdefghijklmnopqrstuvwxyz')
num_array = np.arange(26)
dict = dict(zip(char_list, num_array))

# Creating pandas Series
series_from_list = pd.Series(char_list)
series_from_array = pd.Series(num_array)
series_from_dict = pd.Series(dict)

print(series_from_list)
print(series_from_array)
print(series_from_dict)
```

**W.4 Pandas Practice con**

Compute the minimum, 25th percentile, median, 75th, and maximum of the following series:

ser = pd.Series(np.random.normal(10, 5, 25))

What kind of distribution is this?


```python
import matplotlib.pyplot as plt
ser = pd.Series(np.random.normal(10, 5, 25))
print(ser)
# Compute the required statistics
min_value = ser.min()
percentile_25 = ser.quantile(0.25)
median = ser.median()
percentile_75 = ser.quantile(0.75)
max_value = ser.max()
print(min_value, percentile_25, median, percentile_75, max_value)

ser_resort = ser.sort_values()
ser_resort = ser_resort.reset_index(drop=True)
plt.hint(ser_resort)
plt.show()
```
