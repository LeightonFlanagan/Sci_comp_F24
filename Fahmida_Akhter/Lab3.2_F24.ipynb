```python
#Uploading data
Jefferson_Parish = pd.read_csv("Jefferson_Tract_Coastal.csv")
```

```python
Jefferson_Parish
```

```python
#subseting the table 
#my_dataframe[subset_here].table() method

JP= Jefferson_Parish[["Tract", "Renter_Occupied_House", "Owner_Occupied_House", "Median_Income", "Black_Population", "White_Population"]]
```

```python
JP
```

```python
#Plotting the data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Step 1: Define income ranges (bins) and group the data
income_bins = [0, 20000, 40000, 60000, 80000, 100000]
income_labels = ['0-20k', '20-40k', '40-60k', '60-80k', '80-100k']
JP['Income_Group'] = pd.cut(JP['Median_Income'], bins=income_bins, labels=income_labels)

# Step 2: Group data by income group and sum renter/owner occupied houses
grouped = JP.groupby('Income_Group')[['Renter_Occupied_House', 'Owner_Occupied_House']].sum()

# Step 3: Create the grouped bar chart
bar_width = 0.35
index = np.arange(len(grouped))

renter_color = 'blue'
owner_color = 'salmon'

plt.bar(index, grouped['Renter_Occupied_House'], bar_width, label='Renter Occupied', color=renter_color)
plt.bar(index + bar_width, grouped['Owner_Occupied_House'], bar_width, label='Owner Occupied', color=owner_color)

# Step 4: Add labels, title, and ticks
plt.xlabel('Income Group')
plt.ylabel('Number of Households')
plt.title('Renter vs Owner Occupied Households by Income Group')
plt.xticks(index + bar_width / 2, grouped.index)  # Set the tick labels to income groups
plt.legend()

# Display the plot
plt.show()
```

```python
# Step 2: Group data by income group and sum renter/owner occupied houses
grouped = JP.groupby('Income_Group')[['Black_Population', 'White_Population']].sum()

# Step 3: Create the grouped bar chart
bar_width = 0.35
index = np.arange(len(grouped))

B_color = 'Green'
W_color = 'Brown'

plt.bar(index, grouped['Black_Population'], bar_width, label='Black', color=B_color)
plt.bar(index + bar_width, grouped['White_Population'], bar_width, label='White', color=W_color)


# Step 4: Add labels, title, and ticks
plt.xlabel('Income Group')
plt.ylabel('Population')
plt.title('Black vs White Population by Income Group')
plt.xticks(index + bar_width / 2, grouped.index)  # Set the tick labels to income groups
plt.legend()

# Display the plot
plt.show()
```
