E4: I am trying to make a boxplot here. To do that, we created some of the variables including four types of vegetables
I created a list of font style including color, size  etc.

```python
np.random.seed(19680801)
fruit_weights = [
    np.random.normal(130, 10, size=100),
    np.random.normal(125, 20, size=100),
    np.random.normal(120, 30, size=100),
    np.random.normal(145, 40, size=100)
]
font = {'family': 'serif',
        'color':  "black",
        'weight': 'normal',
        'size': 12,
        }
labels = ['peaches', 'oranges', 'tomatoes','potatoes']
colors = ['peachpuff', 'orange', 'tomato', 'brown'] 
```

```python
In the following code, I set the xaxis, yaxis, title and fill the each boxplot with the colors. 
```

```python
fig, ax=plt.subplots()
bplot=ax.boxplot(fruit_weights,
                  patch_artist=True,
                  labels=labels)  
                     
ax.set_xlabel("Vegetable Types",fontdict= font)
ax.set_ylabel("Fruit Weight in g", fontdict= font)
ax.set_title("Boxplot of fruit weight according to vegetable types", fontdict=font)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)


plt.show()
```

```python
Categorical variable with frequency or counts can be shown effectively through boxplot
```

E6: In this exercise, I created a pie chart for each of the categories of different types of animals. By using the following code and argument" explode" we can explode a section and we can use a customize color pallet as well. Other than that, we can use the argument "autopct='%1.1f%%' " to show the percentage of each slice in the pie chart. In addition we can create white border line around the slices by using the argument "wedgeprops={'edgecolor': 'white'}"

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.1, 0)
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode,
       colors=colors, shadow=True, wedgeprops={'edgecolor': 'white'})
```

```python
In the following code, I tried to create a animation of a heat map by using a mathematical function that takes x and y as inputs and returns the sum of the sine of x and the cosine of y.
x and y are arrays of linearly spaced values between 0 and 2π. These arrays represent the grid of points on which the function f(x, y) will be evaluated.
x is a 1D array with 120 points, and y is reshaped into a 2D column vector with 100 points. This ensures that x and y can be broadcast together when passed into f(x, y).
interval=50: Each frame is shown for 50 milliseconds, resulting in a 20 frames per second (FPS) animation.
blit=True: This optimizes the animation by only redrawing the parts of the image that change between frames.
repeat_delay=1000: The animation will pause for 1 second after completing a loop before starting again. We can customize all of these. We can save it as well. 
```

```python
fig, ax = plt.subplots()
def f(x, y):
    return np.sin(x) + np.cos(y)
x = np.linspace(0, 2 * np.pi, 120)
y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
cmap = "twilight"
ims = []
for i in range(60):
    x += np.pi / 15
    y += np.pi / 30
    im = ax.imshow(f(x, y), animated=True, cmap=cmap)
    if i == 0:
        ax.imshow(f(x, y), cmap=cmap)
        ims.append([im])
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=1000)

writer = PillowWriter(fps=15)
ani.save("movie2.gif", writer=writer)
plt.show()
```

```python
num_arcs = 15
theta = np.linspace(0, np.pi, 36)
radii = np.linspace(4, 5.5, num=num_arcs)
arcs = [np.column_stack([r * np.cos(theta), r * np.sin(theta)]) for r in radii]

fig, ax = plt.subplots(figsize=(6.4, 3))
# set axes limits manually because Collections do not take part in autoscaling
ax.set_xlim(-6, 6)
ax.set_ylim(0, 6)
ax.set_aspect("equal")  # to make the arcs look circular

# create a LineCollection with the half-circles and color mapping
line_collection = LineCollection(arcs, array=radii, cmap="rainbow")
ax.add_collection(line_collection)

fig.colorbar(line_collection, label="Radius")
ax.set_title("Line Collection with mapped colors")

plt.show()
```
