### Box plots

#### Example 1

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(18690801)
fruit_weights = [
    np.random.normal(130, 10, size = 100),
    np.random.normal(125, 20, size = 100),
    np.random.normal(120, 30, size = 100),
]
labels = ['mango', 'avocado', 'guava']
colors = ['yellow', 'green', 'brown']

fig, ax = plt.subplots()
ax.set_ylabel('fruit weight(g)')

bplot = ax.boxplot(fruit_weights,
                patch_artist = True,
                labels = labels)
                #(tick_labels = labels) from the manual gives an error.  
#Note that without this, the colors are not filed
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.show();
```

#### Example 2

```python
np.random.seed(18690801)
family_weights = [
    np.random.normal(120, 10, size = 90),
    np.random.normal(115, 15, size = 100),
    np.random.normal(110, 20, size = 100),
]   
labels = ['Kesse', 'Kojo', 'Abena']
colors = ['black', 'blue', 'Yellow']

fig, ax = plt.subplots()
ax.set_ylabel('family weight(g)')

bplot = ax.boxplot(family_weights,
                patch_artist = True,
                labels = labels)
                
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.show();
```

### Confidence bands

#### Example 1

```python
N = 21
x = np.linspace(0, 16, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1,  9.9, 13.9, 15.1, 12.5]

#fit a linear curve and estimate its y-values & their error
a, b = np.polyfit(x, y, deg = 1)
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) + 
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))

fig, ax = plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha = 0.2)  #alpha determines how deep the error fill is.
ax.plot(x, y, 'o', color = 'tab:green');
```

### Error Function

#### Example 1

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
fig, ax = plt.subplots()
ax.errorbar(x, y, xerr = 0.2, yerr = 0.4)
plt.show()
```

#### Example 2

```python
x = np.arange(0.4, 10, 0.8)
y = np.exp(-x)
fig, ax = plt.subplots()
ax.errorbar(x, y, xerr = 0.1, yerr = 0.04) #xerr and yerr determines how long the error bars are for the x and y axis respectively
plt.show()
```
