**E.3**: Violin plot. Violin plots are useful to represent the data density/dispersion (along the y axis) across different groups and where the mayority of the data falls into when doing statistical analysis (how far the datapoints are from the mean/median values for each group).

```python
#Original plot
# Import the package
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"]=5,5

#Define the style of the plot (see options at https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
plt.style.use('_mpl-gallery')

#make data: using np.random.normal that helps draw random samples  from a normal Gaussian distribution
np.random.seed(10)
D = np.random.normal((3,5,4),(0.75, 1.00, 0.75), (200, 3))

#Create the plot and subplot groups
fig, ax = plt.subplots()
# create the violin plot by defining the dataset, positions of the subplots, and the width of each plot
vp = ax.violinplot(D, [2,4,6], widths =2, showmeans = False, showmedians = False, showextrema = False)

#adjust style
for body in vp['bodies']:
    body.set_alpha(0.9)
ax.set(xlim = (0,8), xticks = np.arange(1,8),
       ylim= (0,8), yticks = np.arange(1,8))

# See plot
plt.show()
```

**E.6**: Scatter plot with Filling the area between lines. Scatter plots are useful to show the relationship between two variables, particularly useful when plotting the statistical relationship and showing both the observations, as well as the regression lines with the confidence intervals.

```python
#Original plot
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"]=7,7

#create the data
N = 21
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1,  9.9, 13.9, 15.1, 12.5]

# fit a linear curve and estimate its y-values and their error.
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) +
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))

#Create the plot
fig, ax = plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)
ax.plot(x, y, 'o', color='tab:brown')
```

**E.7** Nested grids. Useful to show multiple plots at once. In this case I am using regular line plots as example.

```python
#Original example

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=5,5

def example_plot(ax):
    ax.plot([1, 2])
    ax.set_xlabel('x-label', fontsize=12)
    ax.set_ylabel('y-label', fontsize=12)
    ax.set_title('Title', fontsize=14)
    
fig, axs = plt.subplots(nrows=1, ncols=2, layout='constrained') #constrained asjust labels automatically to reduce overlap

for ax in axs.flat:
    example_plot(ax)
```

**E.8**: Boxplot. Usefull to show data dispersion from the mean/median among groups.

```python
# Original example
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"]=5,5

plt.style.use('_mpl-gallery')
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# plot
fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
