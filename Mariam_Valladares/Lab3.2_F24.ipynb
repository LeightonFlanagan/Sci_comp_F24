```python
## 19.3 plot the data nutrients vs streamflow
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=5,5

## organizing the dataset
site1 = nut_sim.loc[['USGS-50010500'],['Year', 'load_kgyear', 'mean_Q_cms']]
site2 = nut_sim.loc[['USGS-50049100'],['Year', 'load_kgyear', 'mean_Q_cms']]
site3 = nut_sim.loc[['USGS-50046000'],['Year', 'load_kgyear', 'mean_Q_cms']]
site4 = nut_sim.loc[['USGS-50047820'],['Year', 'load_kgyear', 'mean_Q_cms']]

site1.index = site1['Year']
site2.index = site2['Year']
site3.index = site3['Year']
site4.index = site4['Year']

site1l = site1.loc[:,['load_kgyear']]
site1q = site1.loc[:,['mean_Q_cms']]
site2l = site2.loc[:,['load_kgyear']]
site2q = site2.loc[:,['mean_Q_cms']]
site3l = site3.loc[:,['load_kgyear']]
site3q = site3.loc[:,['mean_Q_cms']]
site4l = site4.loc[:,['load_kgyear']]
site4q = site4.loc[:,['mean_Q_cms']]

#Create the plot
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex = True, sharey = True)
ax1.scatter(np.log(site1q), np.log(site1l), label = 'USGS-50010500', color = 'lightblue')
ax2.scatter(np.log(site2q), np.log(site2l), label = 'USGS-50049100', color = 'lightgreen')
ax3.scatter(np.log(site3q), np.log(site3l), label = 'USGS-50046000', color = 'red')
ax4.scatter(np.log(site4q), np.log(site4l), label = 'USGS-50047820',color = 'orange')

## Adding annotations
ax1.annotate('USGS-50010500',
            xy=(1, 1), xycoords='data',
            xytext=(10, 10), textcoords='offset points',
            horizontalalignment='left', verticalalignment='bottom')
ax2.annotate('USGS-50049100',
            xy=(1, 1), xycoords='data',
            xytext=(10, 10), textcoords='offset points',
            horizontalalignment='left', verticalalignment='bottom')
ax3.annotate('USGS-50046000',
            xy=(1, 1), xycoords='data',
            xytext=(10, 10), textcoords='offset points',
            horizontalalignment='left', verticalalignment='bottom')
ax4.annotate('USGS-50047820',
            xy=(1, 1), xycoords='data',
            xytext=(10, 10), textcoords='offset points',
            horizontalalignment='left', verticalalignment='bottom')


#format labels
fig.suptitle('Nutrient Exports Puerto Rico Watersheds 1950-2020')
fig.supxlabel('Stream Flow (cms)')
fig.supylabel('Phosphorus loads (kg year)')

plt.tight_layout()
#plt.legend()
plt.plot()
```

## Summary

**Data Methods and Results** 
**Methods:** This brief study summarizes phosphorus loadings for four watersheds in Puerto Rico, using data from water quality monitoring stations available on the Water Quality Portal. The data, which spans from 1950 to 2020, includes nutrient concentrations (g/day). We assessed each site for data gaps and missing observations. Given that all sites had significant data gaps, we first examined the relationship between nutrient concentrations and streamflow to determine an appropriate method for predicting these gaps. A random forest algorithm was employed to train models and predict missing data for each site. Summary statistics, including mean, median, maximum, and minimum values, were then calculated to characterize the nutrient concentration and streamflow datasets.

**Results:** The phosphorus loadings for the studied watersheds showed a weak correlation with discharge (Figure 1), suggesting that traditional methods like rating curve regression may be ineffective for these sites. The random forest approach tended to overestimate nutrient loadings (Figure 2 and 3), particularly during periods with extensive data gaps, such as in the USGS-50047820 watershed. Among the watersheds, USGS-50047820 exhibited the highest mean and median streamflow values, while USGS-50046000 had the highest mean phosphorus loading values.
