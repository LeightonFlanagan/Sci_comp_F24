```python
#In Python, a dictionary key is a unique identifier used to access and store values in a dictionary. Keys are:
#1. Immutable (cannot be changed once created)
#2. Unique (no duplicate keys allowed)
#3. Hashable (can be used as dictionary keys)

#Changing Key Names:

#Key names cannot be directly changed once created. However, you can:
#1. Create a new key with the corrected name and assign the old key's value to it.
#2. Delete the old key.

# Initial dictionary
family_friends = {
    "Ohondwa": {"age": 37, "hairstyle": "locks", "pet": "none"},
    "Chijioke": {"age": 41, "hairstyle": "short", "pet": "none"},
    }

# Print initial dictionary
print("Initial Dictionary:")
print(family_friends)

# Correcting family member's name (Jhon -> John)
family_friends["Owhonda"] = family_friends.pop("Ohondwa")

# Correcting family member's pet type (dog -> golden retriever)
family_friends["Owhonda"]["pet"] = "pitbull"

# Print updated dictionary
print("\nUpdated Dictionary:")
print(family_friends)




```

**E.13** What is the difference between the loc and iloc methods? Explain

```python
#loc (label based) is a technique to select your data based on label when iloc (integer position based) is based on position

#loc (Label-Based Indexing)
#Example:

df.loc['Owhonda', 'age']  # Access age of Owhonda
df.loc['Owhonda':'Kesse', 'age':'hairstyle']  # Access age and hairstyle for Owhonda to Kesse

#iloc (Integer-Based Indexing)

#Example:
df.iloc[0, 1]  # Access age of first row (Owhonda)
df.iloc[0:2, 0:2]  # Access first two rows and columns
```

**E.14** Now demonstrate. Print out both the hairstyle and pet type of your lab partner together (in the same line of code), first using the loc method and then with the iloc method.

```python

import pandas as pd

# Create DataFrame
data = {
    "Name": ["Owhonda", "Chijioke", "Kesse", "Kolin Tyler Bilbrew", "Henry"],
    "age": [37, 41, 24, 22, 31],
    "hairstyle": ["locks", "short", "short", "", "short"],
    "pet": ["none", "none", "none", "Dog", "none"]
}
df = pd.DataFrame(data).set_index("Name")

# Print DataFrame
print(df)

# Using loc method
print("Loc method:")
print(df.loc["Kolin Tyler Bilbrew", ["hairstyle", "pet"]])

# Using iloc method
print("\nIlloc method:")
print(df.iloc[3, [1, 2]])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into DataFrame
df = pd.read_csv('annual_aqi_by_county_2019.csv')

print(df)

subset = df.nsmallest(10, 'Max AQI')

table = subset[['State', 'County', 'Year', 'Days with AQI', 'Good Days', 'Max AQI', 'Median AQI',  'Days CO',  'Days NO2',  'Days Ozone',  'Days PM2.5',  'Days PM10']].copy()
caption = 'Table 1: Top 10 Counties with Worst Air Quality Index (AQI) in 2019'

# Print the table and caption
print(caption)
print(table)
```

```python
columns = ['State', 'County', 'Year', 'Days with AQI', 'Good Days', 
           'Max AQI', 'Median AQI', 'Days CO', 'Days NO2', 
           'Days Ozone', 'Days PM2.5', 'Days PM10']

# Create a figure with subplots
fig, axs = plt.subplots(3, 4, figsize=(20, 15))

# Plot each column
for i, column in enumerate(columns[2:]):
    axs[i // 4, i % 4].scatter(df['State'], df[column])
    axs[i // 4, i % 4].set_title(column)
    axs[i // 4, i % 4].set_xlabel('State')
    axs[i // 4, i % 4].set_ylabel(column)

# Hide unused subplots
for i in range(len(columns[2:]), 12):
    axs[i // 4, i % 4].axis('off')

# Layout so plots do not overlap
fig.tight_layout()

# Figure caption
caption = 'Figure 1: Air Quality Metrics by State and County in 2019'

# Print the caption
print(caption)

# Display the plot
plt.show()
```

```python
# Basic statistics
print("Basic Statistics:")
print(df.describe())

# Maximum values
print("\nMaximum Values:")
print(df[['Days with AQI', 'Good Days', 'Max AQI', 'Median AQI', 'Days CO', 'Days NO2', 'Days Ozone', 'Days PM2.5', 'Days PM10']].max())

# Minimum values
print("\nMinimum Values:")
print(df[['Days with AQI', 'Good Days', 'Max AQI', 'Median AQI', 'Days CO', 'Days NO2', 'Days Ozone', 'Days PM2.5', 'Days PM10']].min())

# Standard deviation
print("\nStandard Deviation:")
print(df[['Days with AQI', 'Good Days', 'Max AQI', 'Median AQI', 'Days CO', 'Days NO2', 'Days Ozone', 'Days PM2.5', 'Days PM10']].std())

# Correlation between columns
print("\nCorrelation between columns:")
print(df[['Days with AQI', 'Good Days', 'Max AQI', 'Median AQI', 'Days CO', 'Days NO2', 'Days Ozone', 'Days PM2.5', 'Days PM10']].corr())

# Top 5 counties with worst air quality (Max AQI)
print("\nTop 5 counties with worst air quality (Max AQI):")
print(df.nlargest(5, 'Max AQI')[['County', 'State', 'Max AQI']])

# Top 5 counties with best air quality (Max AQI)
print("\nTop 5 counties with best air quality (Max AQI):")
print(df.nsmallest(5, 'Max AQI')[['County', 'State', 'Max AQI']])
```
