Create a program that asks the user for a number and then prints out a list of all the divisors of that number. (If you don’t know what a divisor is, it is a number that divides evenly into another number. For example, 13 is a divisor of 26 because 26 / 13 has no remainder.)

```python
def find_divisors():
    num = int(input("Enter a number: "))

    # Initialize an empty list to store divisors
    divisors = []

    # Iterate from 1 to the number
    for i in range(1, num + 1):
        # Check if the number is divisible by i
        if num % i == 0:
            # If divisible, add i to the divisors list
            divisors.append(i)

    # Print the divisors
    print(f"The divisors of {num} are: {divisors}")


find_divisors()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def create_aqi_subset(file_path):
    """
    Creates a subset of the data with the specified columns: State, Days with AQI, Days Ozone, and Days PM2.5.

    Args:
        file_path (str): Path to the CSV file.

    Returns:
        pd.DataFrame: Subset of the data containing the specified columns.
    """
    # Load the data into a DataFrame
    my_dataframe = pd.read_csv(file_path)

    # Display the column names to identify the correct columns (optional)
    print("Column names in the DataFrame:", my_dataframe.columns.tolist())

    # Define the subset columns based on correct column names observed in the dataset
    subset_columns = ['State', 'Days with AQI', 'Days Ozone', 'Days PM2.5']

    # Create the subset of the DataFrame
    try:
        subset = my_dataframe[subset_columns]
        print(subset.head())  # Display the first few rows of the subset (optional)
        return subset
    except KeyError as e:
        print(f"Error: {e}. Please ensure the column names match exactly with the dataset.")
        return pd.DataFrame()  # Return an empty DataFrame on error


def combined_plot(subset_data):
    """
    Combines a line graph of average days with AQI, Ozone, and PM2.5 with a bar chart of average Ozone and PM2.5 pollution days by state.
    
    Args:
        subset_data (pd.DataFrame): Subset of the data containing State, Days with AQI, Days Ozone, and Days PM2.5.
    """
    # Calculate the average number of days by state
    avg_days_by_state = subset_data.groupby('State').mean().reset_index()
    states = avg_days_by_state['State']
    
    # Create the figure and GridSpec layout
    fig = plt.figure(figsize=(14, 10))
    gs = gridspec.GridSpec(2, 1, height_ratios=[2, 1])  # Two rows, line graph (2/3) and bar chart (1/3)
    
    # Top Plot: Line Graph
    ax1 = fig.add_subplot(gs[0])
    ax1.plot(states, avg_days_by_state['Days with AQI'], label='Average Days with AQI', marker='o', color='blue')
    ax1.plot(states, avg_days_by_state['Days Ozone'], label='Average Days Ozone', marker='x', color='orange')
    ax1.plot(states, avg_days_by_state['Days PM2.5'], label='Average Days PM2.5', marker='s', color='green')
    ax1.set_ylabel('Average Number of Days')
    ax1.set_title('Average Days with AQI, Ozone, and PM2.5 by State')
    ax1.legend(loc='upper left')
    ax1.tick_params(axis='x', rotation=90)
    
    # Bottom Plot: Bar Chart
    ax2 = fig.add_subplot(gs[1], sharex=ax1)  # Share the x-axis with the line graph
    ax2.bar(states, avg_days_by_state['Days Ozone'], label='Average Days Ozone', alpha=0.7, color='orange')
    ax2.bar(states, avg_days_by_state['Days PM2.5'], label='Average Days PM2.5', alpha=0.7, color='green')
    ax2.set_xlabel('State')
    ax2.set_ylabel('Average Days of Pollution')
    ax2.legend(loc='upper left')
    ax2.tick_params(axis='x', rotation=90)
    
 
    plt.tight_layout()
    plt.show()


file_path = 'annual_aqi_by_county_2019.csv'
subset_data = create_aqi_subset(file_path)

# Plot the combined graphs
if not subset_data.empty:
    combined_plot(subset_data)
```
