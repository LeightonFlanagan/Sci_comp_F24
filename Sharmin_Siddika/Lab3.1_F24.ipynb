```python
import matplotlib.pyplot as plt
import numpy as np
animals = ("Mouse", "Cat", "Dog", "Goat")
means = {
    'Height': (5.35, 18.43, 25.98, 30.24),
    'Length': (8.79, 38.83, 47.50, 55.45),
    'Teeth': (0.95, 1.82, 2.19, 2.01),
}

x = np.arange(len(animals))  
width = 0.25  
multiplier = 0
fig, ax = plt.subplots()
for attribute, measurement in means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

plt.ylabel('Length (mm)')
ax.set_title('Animal attributes by species')
ax.legend(loc='upper left', ncols=3)
plt.ylim(0, 60)
ax.yaxis.set_minor_locator(mticker.MultipleLocator(2))
plt.show()
```

```python
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2018]
population_countries = {
    'Bangladesh': [.228, .284, .365, .477, .631, .814, 1.044, 1.275],
    'SriLanka': [.340, .425, .519, .619, .727, .840, .943, 1.006],
    'China': [1.394, 1.686, 2.120, 2.625, 3.202, 3.714, 4.169, 4.560],
    'Japan': [.220, .253, .276, .295, .310, .303, .294, .293],
    'Nepal': [.012, .015, .019, .022, .026, .031, .036, .039]}

fig, ax = plt.subplots()
ax.stackplot(year, population_countries.values(),
             labels=population_countries.keys(), alpha=0.65)
ax.legend(loc='upper left')
ax.set_title('World population')
plt.xlabel('Year')
plt.ylabel('Number of people (billions)')
# tick at every 200 million people
ax.yaxis.set_minor_locator(mticker.MultipleLocator(.2))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fruit_weights = [
    [128, 130, 127, 133, 129],  
    [125, 127, 123, 130, 126],  
    [120, 125, 124, 126, 122]   
]
labels = ['mangoes', 'oranges', 'applea']
colors = ['yellow', 'orange', 'red']

fig, ax = plt.subplots()
plt.ylabel('fruit weight (g)')

bplot = ax.boxplot(fruit_weights,
                   patch_artist=True, labels=labels)

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

Useful for showing the distribution of data, including the median, quartiles, and outliers. It provides a clear view of the data spread and central tendency.

```python
import matplotlib.pyplot as plt

# Sample Data
age = [23, 22, 55, 65]
name = ['Sharmin', 'Srestho', 'Jahin', 'Fahim']

# Plotting
plt.pie(age, labels=name)
plt.title('Pie Chart of age')
plt.show()
```

```python
explode = (0.1, 0, 0.2, 0) 
colors = ['yellow', 'coral', 'skyblue', 'lightgreen']
plt.pie(age, labels=name, explode=explode, colors=colors)
plt.title('New Pie Chart')
plt.show()
```

```python
plt.pie(age, labels=name, autopct='%1.1f%%', colors=colors, explode=explode)
plt.title('New Chart with Percentage')
plt.show()
```

Useful for showing proportions or percentages of a whole. Each slice represents a category's contribution to the total.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('X Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Modified Histogram')
plt.xlabel('X Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data, bins=20, density=True, alpha=0.5, color='green', edgecolor='black')
plt.title('Modified Histogram')
plt.xlabel('X Value')
plt.ylabel('Frequency')
plt.show()
```

Useful for showing the distribution of a dataset. It displays how frequently different ranges of values occur, helping in understanding the underlying distribution.
