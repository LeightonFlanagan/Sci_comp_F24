**E.13** What is the difference between the loc and iloc methods? Explain

The loc and iloc methods in pandas are used for accessing data within a DataFrame, but they work differently based on the types of indexing we want to perform:
loc uses labels (names or index labels). iloc uses integer positions.
loc is inclusive of the end labels. iloc is exclusive of the end positions.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Landuse Type': ['Agricultural Land', 
                         'Barren Land', 'Buildup Area', 
                         'Vegetation', 'Waterbody'],
    'Year 2000': [23.37, 8.65, 14.55, 48.86, 4.57],
    'Year 2010': [20.95, 8.81, 18.03, 47.82, 4.39],
    'Year 2020': [17.01, 9.38, 23.19, 44.1, 6.32]}

df = pd.DataFrame(data)
df.set_index('Landuse Type', inplace=True)

ax = df.plot(kind='bar', figsize=(12, 8), width=0.8)

for container in ax.containers:
    ax.bar_label(container, label_type='edge')
    
plt.title('Land Use Land Cover change detection by Year')
plt.xlabel('Landuse Type')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# maximum value for the Year 2020
max_2020 = df['Year 2020'].max()

print("Maximum value for Year 2020:", max_2020)
```

```python
# lowest amount of agricultural land
agricultural_land = df.loc['Agricultural Land']
lowest_amount_year = agricultural_land.idxmin()
lowest_amount_value = agricultural_land.min()

print(f"The year with the lowest amount of Agricultural Land is {lowest_amount_year} with a value of {lowest_amount_value}.")
```
