**W.1** Write a function that returns a list of the first n powers of 2 (2^0, 2^1, 2^2, 2^3...2^n).

```python
def power_2 (n):

    return [2 ** i for i in range(0,n+1)]

print(power_2(4))
```

**W.2** Using the function above, plot the first 100 powers of 2

```python
import matplotlib.pyplot as plt

n = 100

power = power_2(n)

plt.figure(figsize=(10, 6))
plt.plot(power, marker='o', linestyle='-', markersize=4)
plt.yscale('log')  # Use logarithmic scale to better visualize large numbers
plt.title("First 100 Powers of 2")
plt.xlabel("Index (n)")
plt.ylabel("2^n (log scale)")
plt.grid(True)
plt.show()
```
