### Using packages within functions

We can also have import statements within a function, but this can slow the function down if the package is not loaded already

```python
def circle_area6(radius):
    if radius <= 0:
        print('Error: radius needs to be positive')
        return # exit the function
    
    import numpy as np
    
    return np.pi*radius**2
```

```python
circle_area6(3) # note this is a much more accurate
```

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# add csv file here
data_sd = pd. read_csv('alldata_sed.csv',index_col = 'date')
data_sd
```

```python
# Create a function to find max values

def get_max_value(data_sd, element):
    """I want to find max value for each site and the date we measured"""
    result = data_sd.groupby('site').apply(lambda x: x.loc[x[element].idxmax(), [element]])
    result['max_date'] = data_sd.groupby('site').apply(lambda x: x[element].idxmax())
    result.rename(columns={element: 'max_value'}, inplace=True)
    return result

max = get_max_value(data_sd, 'Na')
print(max)
```

```python
def plot_salinity_over_time(data_sd):
    """Plots Salinity_mean over time for each site"""
    
    # Pivot the table to have dates on the x-axis and separate lines for each site
    df_pivot = data_sd.pivot_table(index=data_sd.index, columns='site', values='Salinity_mean')
    
    # Plot
    plt.figure(figsize=(10, 6))
    
    # Plot each site's Salinity_mean
    for site in df_pivot.columns:
        plt.plot(df_pivot.index, df_pivot[site], label=f'Site {site}')
    
    # Adding labels and title
    plt.xlabel('Date')
    plt.ylabel('Salinity Mean')
    plt.title('Salinity Mean Over Time by Site')
    plt.legend(title='Site')
    plt.grid(True)
    
    # Show the plot
    plt.show()

plot_salinity_over_time(data_sd)
```
