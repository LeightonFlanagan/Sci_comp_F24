```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

Drainage_Watershed = ('Vivi', 'Caonillas', 'Carite', 'Cerrillos')
y_pos = np.arange(len(Drainage_Watershed))
num_microcatchments = 100 + 300 * np.random.rand(len(Drainage_Watershed))
error = np.random.rand(len(Drainage_Watershed))

labels = ['Vivi', 'Caonillas', 'Carite', 'Cerrillos']
microcatchments = [20,209,24,79]

fig, ax = plt.subplots()
ax.barh(y_pos, microcatchments, xerr=error, align='center')
ax.set_yticks(y_pos, labels=Drainage_Watershed)
ax.invert_yaxis()
ax.set_xlabel('Number of Microcatchments')
ax.set_title('Puerto Rico SDR')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

locations = ('Vivi', 'Caonillas', 'Cerrillos')
sediment_means = {
    'Total Sediment Exports': (13380.74, 151338.2, 12441.28),
    'Total Soil Loss': (64828.72, 702676.71, 62760.30),
}

x = np.arange(len(locations)) 
width = 0.35 
multiplier = 0

fig, ax = plt.subplots(layout='none')

for attribute, measurement in sediment_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1


ax.set_ylabel('Ton/Year')
ax.set_title('Puerto Rico Sediment Delivery Ratio')
ax.set_xticks(x + width, locations, rotation=45)
ax.legend(loc='center left', fontsize=10, title_fontsize=15, ncols=3)

ax.set_ylim(0, 800000)

plt.show()
```
