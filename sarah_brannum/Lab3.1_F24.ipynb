```python
# E 4
# Violin Plot
# These plots show similar information to a histogram since it shows probability distribution, but are helpful to show multiple variables next to each other

# fake data
fs = 10  # fontsize
pos = [1, 2, 4, 5, 7, 8]
data = [np.random.normal(0, std, size=100) for std in pos] # 6 different arrays with 100 data points per array

fig, axs = plt.subplots()
axs.violinplot(data, pos, points=20, widths=0.3,
                     showmeans=True, showextrema=True, showmedians=True)
axs.set_title('Custom violin 1', fontsize=fs)
```

```python
fig, axs = plt.subplots()
axs.violinplot(data, pos, points=20, widths=0.3,
                     showmeans=True, showextrema=True, showmedians=True)
axs.set_title('Look! The title changed!', fontsize=fs) #change the title
```

```python
fig, axs = plt.subplots()
axs.violinplot(data, pos, points=20, widths=0.3,
                     showmeans=False, showextrema=False, showmedians=False) #showing just the shadow and not the lines on the violin plot
axs.set_title('Custom violin 1', fontsize=fs)
```

```python
# E 5
# Contour plot
# This plot is useful to show a spatially changing variable over a location. An example would be rainfall amounts over Louisiana during Hurricane Francine
```

```python
import matplotlib.cm as cm

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=30) #VERY large labels for the contour lines
ax.set_title('Simplest default with labels')
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, colors='green') #change color of contour plot
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')
```

```python
# E 6
# Bar plot on polar axes
# This plot can show wind direction, similar to a wind rose (there might be a better way to plot a wind rose but this is pretty close)
# A plot like this would be good to show wind direction and magnitude for the year
```

```python
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

plt.show()
```
