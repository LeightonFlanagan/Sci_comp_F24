**W.1** Make a map of Antarctica using a polar projection in cartopy. Add a marker for McMurdo Station.

```python
plt.figure(figsize=(7, 5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-180,180,-90,-50]) # set the limits of the plot
#set xmin, xmax, ymin, ymax
ax.coastlines(resolution='110m')
ax.gridlines(draw_labels=True)
```

```python
plt.figure()
ax=plt.axes(projection=ccrs.SouthPolarStereo())
ax.set_extent([-180,180,-90,-60],ccrs.PlateCarree()) # set the limits of the plot
#set xmin, xmax, ymin, ymax
ax.coastlines(resolution='110m')
ax.gridlines(draw_labels=True)
ax.plot(166,-78,marker='o',transform=ccrs.PlateCarree()) #lon, lat, need transformation
plt.show()
```
